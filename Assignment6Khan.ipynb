```python
import random

bowling = open('bowling.txt','w')

playernum = int(input("Enter number of players: "))
pname = []
total = []

for i in range(playernum):
    newplayer = input("Enter name of player: ")
    pname.append(newplayer)
    total.append(0)

for i in range(10):
    bowling.write("Round "+str(i+1)+"\n")    
    for j in range(playernum):        
        score1 = random.randint(0,10)
        score2 = random.randint(0,(10-score1))
        bowling.write(pname[j])
        bowling.write("\n")
        
        bowling.write(str(score1)+"\n")        
        if score1 == 10:
            bowling.write("/"+"\n")
        else:
            bowling.write(str(score2)+"\n")        
        total[j] = total[j]+int(score1)+int(score2)
        
        if (i == 9):
            bowling.write("Total: "+str(total[j])+"\n")   
bowling.close

bowling = open('bowling.txt','r')
for i in bowling:
    print (i)
    
bowling.close    
            
```
